```python
def eliminar_columnas(df):
    """
    Esta función elimina las columnas que no queremos tener en nuestro dataframe, puesto que no basaremos nuestro análisis en ellas
    Args: df
    df(dataframe):dataframe al que queremos aplicar la función
    """

    columnas_eliminadas = []
    patron = 'Q4|Q6|Q8|Q11|Q13|Q15|Q20|Q21|Q22|Q26|Q33|Q12|Q16|Q17|Q32|Q34|time|Q23|Q35'

    for col in df.columns:
        if re.findall(patron,col):
            columnas_eliminadas.append(col)
            
    eliminado = df.drop(columnas_eliminadas, axis=1, inplace = True)
    return eliminado
```

# LIMPIEZA: 


Para la columna 'country':

- cambiar el valor 'Iran, Islamic Republic of...' por 'Islamic Republic of Iran'

- el valor 'Viet Nam' está mal escrito, ya que se escribe junto

```python
def limpiar_country(columna):
    """
    Esta función nos permite modificar los datos de una columna para que estén escritos adecuadamente
    Args: columna
    columna(string): es el dato que queremos modificar 
    """
    
    if columna == 'Iran, Islamic Republic of...':
        cambiado = columna.replace('Iran, Islamic Republic of...', 'Islamic Republic of Iran')
        return cambiado
    
    elif columna == 'Viet Nam':
        cambiado2 = columna.replace('Viet Nam', 'Vietnam')
        return cambiado2
    
    else: 
        return columna
```

```python
df['country'] = df['country'].apply(limpiar_country)
df['country'].unique()
```

Hemos aplicado un unique para comprobar que los cambios se han realizado correctamente

Para la columna 'yearly_compensation': 
-  símbolos extraños a eliminar: `$`    

```python
def limpiar_yearly(columna):
    
    """
    Esta función nos permite eliminar el símbolo $ de los datos de nuestras columnas
    Args: columna
    columna(string): es el dato que queremos modificar 
    """
    try:
        return re.sub('\$', '', columna)
         
    except:
        return columna
    
```

```python
df['yearly_compensation'] = df['yearly_compensation'].apply(limpiar_yearly)
df['yearly_compensation'].unique()
```

Hemos aplicado un unique para comprobar que los cambios se han realizado correctamente

Para la columna 'activities':
- eliminar espacios al inicio de algunos strings

```python
def limpiar_activities(columna):
    """
    Esta función nos permite eliminar los espacios que hay al principio y al final de los strings 
    Args: columna
    columna(string): es el dato al que le queremos quitar los espacios 
    """
    try:
        return columna.strip()
         
    except:
        return columna
```

```python
df['activities'] = df['activities'].apply(limpiar_activities)
df['activities'].unique()
```

Hemos aplicado un unique para comprobar que los cambios se han realizado correctamente

Empezamos a categorizar nuestras columnas. Para ello creamos una función 

```python
def rango_edad(col):
    
    joven = ['25-29', '18-21', '22-24']
    mediana =['30-34', '35-39', '40-44', '45-49', '50-54']
    mayor = ['55-59','60-69', '70+']
    
    if col in joven:
        return 'joven'
        
    elif col in mediana:
        return 'mediana'
        
    elif col in mayor:
        return 'mayor'
```

```python
df_analyst['ages'] = df_analyst['age'].apply(rango_edad)
```
