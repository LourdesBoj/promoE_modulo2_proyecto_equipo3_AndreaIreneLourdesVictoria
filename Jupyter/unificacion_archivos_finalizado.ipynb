### Unificamos las columnas que pertenecen a la misma pregunta

Creamos dos funciones ('union' y 'nulos') para cada nueva columna que queramos crear uniendo varias columnas ya existentes. Una vez creada la nueva columna, borraremos las columnas que contenían la información por separado

```python
def union(df, nombre, *args):
  """
    La función combinará los valores que hay en las diferentes columnas y creará una nueva columna que unifique todos esos valores
   Args: 
        df (dataframe) : será el df sobre el que queremos trabajar
        nombre(string): será el nombre de nuestra nueva columna
        *args: serán las columnas que queremos unificar en una sola
        """
  parametros = args
  columnas = list(parametros) # creamos la variable 'columnas' que será una lista con los nombres las columnas.  
     
  
  if len(columnas) == 7:
    df[nombre] = df[columnas[0]].astype(str) + ";" + df[columnas[1]].astype(str) + ";" + df[columnas[2]].astype(str) + ";" + df[columnas[3]].astype(str) + ";" + df[columnas[4]].astype(str) + ";" + df[columnas[5]].astype(str) + ";" + df[columnas[6]].astype(str)
    df[nombre] = df[nombre].str.replace('nan;', '')
    df[nombre] = df[nombre].str.replace(';nan', '')
  

  
  elif len(columnas) == 6:
    df[nombre] = df[columnas[0]].astype(str) + ";" + df[columnas[1]].astype(str) + ";" + df[columnas[2]].astype(str) + ";" + df[columnas[3]].astype(str) + ";" + df[columnas[4]].astype(str) + ";" + df[columnas[5]].astype(str) 
    df[nombre] = df[nombre].str.replace('nan;', '')
    df[nombre] = df[nombre].str.replace(';nan', '')
    
  
  elif len(columnas) == 12:
    df[nombre] = df[columnas[0]].astype(str) + ";" + df[columnas[1]].astype(str) + ";" + df[columnas[2]].astype(str) + ";" + df[columnas[3]].astype(str) + ";" + df[columnas[4]].astype(str) + ";" + df[columnas[5]].astype(str) + ";" + df[columnas[6]].astype(str) + ";" + df[columnas[7]].astype(str) + ";" + df[columnas[8]].astype(str) + ";" + df[columnas[9]].astype(str) + ";" + df[columnas[10]].astype(str) + ";" + df[columnas[11]].astype(str) 
    df[nombre] = df[nombre].str.replace(';nan', '')
    df[nombre] = df[nombre].str.replace(';nan', '')
    
  
  elif len(columnas) == 5:
    df[nombre] = df[columnas[0]].astype(str) + ";" + df[columnas[1]].astype(str) + ";" + df[columnas[2]].astype(str) + ";" + df[columnas[3]].astype(str) + ";" + df[columnas[4]].astype(str) 
    df[nombre] = df[nombre].str.replace('nan;', '')
    df[nombre] = df[nombre].str.replace(';nan', '')
  
  
  if len(columnas) == 8:
    df[nombre] = df[columnas[0]].astype(str) + ";" + df[columnas[1]].astype(str) + ";" + df[columnas[2]].astype(str) + ";" + df[columnas[3]].astype(str) + ";" + df[columnas[4]].astype(str) + ";" + df[columnas[5]].astype(str) + ";" + df[columnas[6]].astype(str) + ";" + df[columnas[7]].astype(str)
    df[nombre] = df[nombre].str.replace('nan;', '')
    df[nombre] = df[nombre].str.replace(';nan', '')
    
    
  elif len(columnas) == 10:
    df[nombre] = df[columnas[0]].astype(str) + ";" + df[columnas[1]].astype(str) + ";" + df[columnas[2]].astype(str) + ";" + df[columnas[3]].astype(str) + ";" + df[columnas[4]].astype(str) + ";" + df[columnas[5]].astype(str) + ";" + df[columnas[6]].astype(str) + ";" + df[columnas[7]].astype(str) + ";" + df[columnas[8]].astype(str) + ";" + df[columnas[9]].astype(str)
    df[nombre] = df[nombre].str.replace(';nan', '')
    df[nombre] = df[nombre].str.replace(';nan', '')
    
    
  elif len(columnas) == 21:
    df[nombre] = df[columnas[0]].astype(str) + ";" + df[columnas[1]].astype(str) + ";" + df[columnas[2]].astype(str) + ";" + df[columnas[3]].astype(str) + ";" + df[columnas[4]].astype(str) + ";" + df[columnas[5]].astype(str) + ";" + df[columnas[6]].astype(str) + ";" + df[columnas[7]].astype(str) + ";" + df[columnas[8]].astype(str) + ";" + df[columnas[9]].astype(str) + ";" + df[columnas[10]].astype(str) + ";" + df[columnas[11]].astype(str) + ";" + df[columnas[12]].astype(str) + ";" + df[columnas[13]].astype(str) + ";" + df[columnas[14]].astype(str) + ";" + df[columnas[15]].astype(str) + ";" + df[columnas[16]].astype(str) + ";" + df[columnas[17]].astype(str) + ";" + df[columnas[18]].astype(str) + ";" + df[columnas[19]].astype(str) + ";" + df[columnas[19]].astype(str)
    df[nombre] = df[nombre].str.replace(';nan', '')
    df[nombre] = df[nombre].str.replace(';nan', '')
    
    
  elif len(columnas) == 17:
    df[nombre] = df[columnas[0]].astype(str) + ";" + df[columnas[1]].astype(str) + ";" + df[columnas[2]].astype(str) + ";" + df[columnas[3]].astype(str) + ";" + df[columnas[4]].astype(str) + ";" + df[columnas[5]].astype(str) + ";" + df[columnas[6]].astype(str) + ";" + df[columnas[7]].astype(str) + ";" + df[columnas[8]].astype(str) + ";" + df[columnas[9]].astype(str) + ";" + df[columnas[10]].astype(str) + ";" + df[columnas[11]].astype(str) + df[columnas[12]].astype(str) + ";" + df[columnas[13]].astype(str) + ";" + df[columnas[14]].astype(str) + ";" + df[columnas[15]].astype(str) + ";" + df[columnas[16]].astype(str)
    df[nombre] = df[nombre].str.replace(';nan', '')
    df[nombre] = df[nombre].str.replace(';nan', '')
    
```

```python
#creamos una función que nos convierte el string 'nan' en un NaN de numpy
def nulos(elemento):
    """
    La función cambia el tipo de dato, de string a nulo
   Args: 
        elemento: será el elemento al que le cambiaremos el tipo de dato
    
   Returns: 
        Si la condición se cumple, la función devolverá el dato cambiado a nulo. Si no se cumple, devolverá el elemento como estaba inicialmente
    """
    if elemento ==  'nan': 
        return np.nan
    else:
        return elemento
```
