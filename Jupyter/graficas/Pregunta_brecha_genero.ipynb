## ¿Existe brecha de género en el sector Data?

```python
df_brecha = df[(df['gender'] == 'Man') | (df['gender'] == 'Woman')]
```

```python
ordenado = df_brecha.sort_values(by='gender', ascending=False)
```

```python
sns.countplot(x = 'ages' , data = df_brecha, hue = 'gender', order = ['joven', 'mediana', 'mayor']);
```

```python
plt.figure(figsize =(10,5))
sns.countplot(x = "ages", data = df_brecha, hue = 'gender', order = ['joven', 'mediana', 'mayor'], palette="crest")
plt.title("Edades según el género", color = "pink", fontweight='bold', fontsize = 16)
plt.xlabel("ides", color = "#3a6670")
plt.ylabel("count", color = "#3a6670")
#fig.tight_layout()
;
```
